```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv/data.csv", encoding="latin1")
data.head(10)
```

```python
data = data.dropna(subset=["no2"])
x = data["no2"].values.astype(float)
print(f"Total samples used: {len(x)}")
```

```python
r = 102316041
ar = 0.05 * (r%7)
br = 0.3 * (r% 5+1)
print(f"ar = {ar}")
print(f"br = {br}")
```

```python
z = x + ar*np.sin(br*x)
```

### PDF: p̂(z) = c * exp(-λ (z - μ)^2)


```python
mu = np.mean(z)
var = np.mean((z-mu) ** 2)
lam = 1/(2*var)
# Normalization const
c = np.sqrt(lam / np.pi)
```

```python
print(f"μ (mean)       = {mu}")
print(f"var (variance) = {mu}")
print(f"λ (lambda)     = {lam}")
print(f"c (constant)   = {c}")
```
